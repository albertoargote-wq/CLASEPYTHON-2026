# **4. Área de un Triángulo**
● Objetivo: Calcular el área de un triángulo
● Conceptos: Fórmulas matemáticas, operaciones con flotantes

```python
base = float (input("ingresa la base del triangulo"))
altura = float (input("ingresa la altura del triangulo"))
area = (base*altura)/2
print("el area del triangulo es =",area)
```

# **5. Conversión de Unidades de Longitud**
● Objetivo: Convertir centímetros a pies y pulgadas
● Conceptos: Conversión de unidades, formato de números decimales

```python
centimetros = float (input("ingresa tus centimetros a convertir"))
pulgadas = centimetros*0.39
pies = centimetros/30.48
print("la conversion a pulgadas es =",pulgadas)
print("la conversion a pies es =",pies)
```

**6. Conversión de Temperaturas**
● Objetivo: Convertir temperatura entre diferentes escalas
● Conceptos: Fórmulas de conversión, operaciones con decimales

```python
centigrados = float (input("ingresa tus grados centigrados a convertir"))
fahrenheit = (centigrados*1.8)+32
kelvin = centigrados + 273.15
print("la conversion en fahrenheit es =",fahrenheit)
print("la conversion en kelvin =",kelvin)
```
